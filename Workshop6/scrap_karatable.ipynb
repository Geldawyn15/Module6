```python
response = requests.get('https://www.kartable.fr/histoire/college').text
```

```python
soup = BeautifulSoup(response, 'lxml')

soup
```

```python
print(soup.prettify())
```

```python
# Find <style> tag
style_content = soup.find('style').get_text()

# Regex to find all font-family declarations
fonts = set(re.findall(r"font-family:'([^']+)'", style_content))

# Printing the set of fonts
print(fonts)
```
