```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
url = "https://www.maths-et-tiques.fr/"
```

```python
page = requests.get(url)
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
tag = soup.find('div', class_ = re.compile(r'custom'))
print(tag)
```
